**Perceptrons**
</br></br>

Perceptrons — invented by Frank Rosenblatt in 1958, are the simplest neural network that consists of $n$ number of inputs, only one neuron, and one output, where $n$ is the number of features of our dataset. The process of passing the data through the neural network is known as forward propagation and the forward propagation carried out in a perceptron is explained in the following three steps.
</br>


**Step 1**: For each input, multiply the input value $x_i$ with weights $w_i$ and sum all the multiplied values. Weights — represent the strength of the connection between neurons and decides how much influence the given input will have on the neuron's output. If the weight $w_1$ has a higher value than the weight $w_2$, then the input $x_1$ will have a higher influence on the output than $w_2$.
</br></br>

$$\sum = (x_1\times w_1) + (x_2\times w_2) + \dots + (x_n\times w_n)$$
</br>

The row vectors of the inputs and weights are $x = [x_1,x_2,\dots,x_n]$ and $w = [w_1,w_2,\dots,w_n]$ respectively and their dot product is given by
</br></br>

$$x\cdot w = (x_1\times w_1) + (x_2\times w_2) + \dots + (x_n\times w_n)$$
</br>

Hence, the summation is equal to the dot product of the vectors x and w
</br></br>

$$\sum = x\cdot w$$
</br>

**Step 2**: Add bias $b$ to the summation of multiplied values and let's call this $z$. Bias — also known as the offset is necessary in most of the cases, to move the entire activation function to the left or right to generate the required output values.
</br></br>

$$z = x\cdot w + b$$
</br>

**Step 3**: Pass the value of $z$ to a non-linear activation function. Activation functions — are used to introduce non-linearity into the output of the neurons, without which the neural network will just be a linear function. Moreover, they have a significant impact on the learning speed of the neural network. Perceptrons have <i>binary step function</i> as their activation function. However, we shall use <i>sigmoid</i> — also known as <i>logistic</i> function as our activation function.
</br></br>

$$\hat{y} = \sigma(z) = \frac{1}{1 + e^{-z}}$$
</br>

$\sigma$ denotes the <i>sigmoid</i> activation function and the output we get after the forward prorogation is known as the <i>predicted value</i> $\hat{y}$.
</br>

**Learning Algorithm**
</br></br>

The learning algorithm consists of two parts — backpropagation and optimization.
</br></br>

**Backpropagation**: Backpropagation, short for <i>backward propagation of errors</i>, refers to the algorithm for computing the gradient of the loss function with respect to the weights. However, the term is often used to refer to the entire learning algorithm. The backpropagation carried out in a perceptron is explained in the following two steps.
</br></br>

**Step 1**: To know an estimation of how far are we from our desired solution a <i>loss function</i> is used. Generally, <i>mean squared error</i> is chosen as the loss function for regression problems and <i>cross entropy</i> for classification problems. Let's take a regression problem and its loss function be mean squared error, which squares the difference between <i>actual</i> ($y_1$) and <i>predicted value</i> ($\hat{y_1}$).
</br></br>

$$MSE_i = (y_i - \hat{y_i})^2$$
</br>

Loss function is calculated for the entire training dataset and their average is called the <i>Cost function</i> $C$.
</br></br>

$$C = MSE = \frac{1}{n}\sum_{i=1}^n (y_i - \hat{y_i})^2$$
</br>



**Step 2:** In order to find the best weights and bias for our Perceptron, we need to know how the cost function changes in relation to weights and bias. This is done with the help of the <i>gradients (rate of change)</i> — how one quantity changes in relation to another quantity. In our case, we need to find the gradient of the cost function with respect to the weights and bias.
</br></br>

Let's calculate the gradient of cost function $C$ with respect to the weight $w_i$ using <i>partial derivation</i>. Since the cost function is not directly related to the weight $w_i$, let's use the [chain rule](https://en.wikipedia.org/wiki/Chain_rule).
</br></br>

$$\frac{\partial C}{\partial w_i} = \frac{\partial C}{\partial \hat{y}} \times \frac{\partial \hat{y}}{\partial z} \times \frac{\partial z}{\partial w_i}$$
</br>

Now we need to find the following three gradients
</br></br>

$$\frac{\partial C}{\partial \hat{y}} = ?,\quad \frac{\partial \hat{y}}{\partial z} = ?,\quad \frac{\partial z}{\partial w_i} = ?$$
</br>

Let's start with the gradient of the <i>cost function</i> ($C$) with respect to the <i>predicted value</i> ($\hat{y}$)
</br></br>

$$\frac{\partial C}{\partial \hat{y}} = \frac{\partial}{\partial \hat{y}}\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y_i})^2 = 2 \times \frac{1}{n}\sum_{i=1}^n (y_i - \hat{y_i})^2$$
</br>

Let $y = [y_1,y_2,\dots,y_n]$ and $\hat{y} = [\hat{y_1},\hat{y_2},\dots,\hat{y_n}]$ be the row vectors of actual and predicted values. Hence the above equation is simplified as
</br></br>

$$\frac{\partial C}{\partial \hat{y}} = \frac{2}{n}\times sum(y - \hat{y})$$
</br>

Now let's find the gradient of the <i>predicted value</i> with respect to the z. This will be a bit lengthy.
</br></br>

\begin{align}
  \frac{\partial \hat{y}}{\partial z} &= \frac{\partial}{\partial z}\sigma(z) \\ \\
  &= \frac{\partial}{\partial z} \left( \frac{1}{1+e^{-z}} \right) \\ \\
  &= \frac{e^{-z}}{(1+e^{-z})^2} \\ \\
  &= \frac{1}{1+e^{-z}}\times \frac{e^{-z}}{1+e^{-z}} \\ \\
  &= \frac{1}{1+e^{-z}}\times \left(1-\frac{1}{1+e^{-z}}\right) \\ \\
  &= \sigma(z)\times (1-\sigma(z))
\end{align}
</br>

The gradient of $z$ with respect to the weight $w_i$ is
</br></br>

\begin{align}
  \frac{\partial z}{\partial w_i} &= \frac{\partial}{\partial w_i}(z) \\ \\
  &= \frac{\partial}{\partial w_i}\sum_{i=1}^n(x_i\cdot w_i + b) \\ \\
  &= x_i
\end{align}
</br>

Therefore we get,
</br></br>

$$\frac{\partial C}{\partial w_i} = \frac{2}{n}\times sum(y-\hat{y})\times \sigma(z)\times (1-\sigma(z))\times x_i$$
</br>

What about Bias? — Bias is theoretically considered to have an input of constant value $1$. Hence,
</br></br>

$$\frac{\partial C}{\partial b} = \frac{2}{n}\times sum(y-\hat{y})\times \sigma(z)\times (1-\sigma(z))$$
</br>


**Optimization**: Optimization is the selection of the best element from some set of available alternatives, which in our case, is the selection of best weights and bias of the perceptron. Let's choose <i>gradient descent</i> as our optimization algorithm, which changes the <i>weights</i> and <i>bias</i>, proportional to the negative of the gradient of the cost function with respect to the corresponding weight or bias. <i>Learning rate</i> ($\alpha$) is a hyperparameter which is used to control how much the weights and bias are changed.
</br></br>

The weights and bias are updated as follows and the backpropagation and gradient descent is repeated until convergence.
</br></br>

\begin{align}
  w_i &= w_i - \left( \alpha\times \frac{\partial C}{\partial w_i} \right) \\ \\
  b &= b - \left( \alpha\times \frac{\partial C}{\partial b} \right)
\end{align}
