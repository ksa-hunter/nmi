```python
import math as m
import numpy as np

def exp(ax):
  y = 0
  for i in range(n): # n to find
    y += pow(ax,i)/m.factorial(i)
  return y

TOL = 1e-4 # epsilon (vs eta if delta-y)
x = 3
y_sys = np.exp(x)
nmax = 25

err = -1
for n in range(nmax):
  y_mod = exp(x)
  err = abs(y_sys-y_mod)
  if err < TOL:
    break

print(f"exp({x})= {y_sys} ~ {y_mod} at {n} steps with error {err} < TOL = {TOL}.")
```
